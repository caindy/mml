```python
%load_ext autoreload
%autoreload 2

from datasets import Zone1
ds = Zone1('data/zone1_prior_load.parquet')
ds.data.head()
```

```python
import statsmodels.api as sm
import numpy as np
from validation import walkforward, Error

class OLSWrapper():
    def fit(self, X, y):
        self.__ols = sm.OLS(y, X)
        self.__fit_results = self.__ols.fit()
        return self

    def predict(self, x):
        prediction = self.__fit_results.predict(x)
        return prediction

ols = OLSWrapper()

(predictions, errors) = walkforward(ols, ds.data, ds.data[ds.actual],
                                    ds.validation_start, ds.validation_end, Zone1.next_hour,
                                    ds.correlated_columns)
overall_error = Error(ds.validation_data[ds.actual], np.concatenate(predictions))
print(overall_error)
```

This is *not* an improvement on the nearest neighbors regression model.

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

pf = PolynomialFeatures(degree=3, interaction_only=True)
class PolynomialOLSWrapper(OLSWrapper):
    def fit(self, X, y):
        y = y.copy()
        X_poly_train = pf.fit_transform(X.copy(), y)
        return super().fit(X_poly_train, y)
    def predict(self, x):
        x = pf.transform(x.copy())
        return super().predict(x)

pls = PolynomialOLSWrapper()

(predictions, errors) = walkforward(pls, ds.data, ds.data[ds.actual],
                                    ds.validation_start, ds.validation_end, Zone1.next_hour,
                                    ds.features)
overall_error = Error(ds.validation_data[ds.actual], np.concatenate(predictions))
print(overall_error)
```
